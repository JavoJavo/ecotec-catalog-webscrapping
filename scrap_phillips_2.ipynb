```python
def concat_df_2dims(dfs):
    dfs_2dims = []
    dfs_with_more_dims_index = []
    for i,df_i in enumerate(dfs):
        if df_i.shape[1] == 2:
            dfs_2dims.append(df_i.reset_index(drop=True))
        else:
            try:
                dfs_2dims.append(uglyt6(df_i).reset_index(drop=True))
            except:
                warnings.warn('Could not process table with 2+ columns')
            dfs_with_more_dims_index.append(i)
    return pd.concat(dfs_2dims).set_index(0).transpose(), dfs_with_more_dims_index
```

```python
def concat_dfs(links):
    concated_dfs = []
    for link in links:
        sleep(random()*2.5)
        try:
            dfs = pd.read_html(link)
        except:
            warnings.warn("Error reading tables. There were no tables or another error ocurred.")
            continue
        c_df, more_than_2_dims = concat_df_2dims(dfs)
        print(more_than_2_dims)
        concated_dfs.append(c_df)
    return pd.concat(concated_dfs).reset_index(drop=True)
```

```python
def links_format(l_limit,u_limit):
    for i in range(l_limit,u_limit):
        print("l{} = ''".format(i))
    print('links = ['+' ,'.join(["l{}".format(i) for i in range(l_limit,u_limit)]) +']')
```

```python
def squeeshingTable6(table6):
    table6 = table6.set_index(0).transpose()
    rows_names = []
    for i in table6.values:
        rows_names.append(str(table6.columns[1])+' | '+'Marca de aprobación: '+i[0])
    values = []
    for v in table6.values:
        values.append(v[1])
    return pd.DataFrame({0:rows_names,1:values})
```

```python
def curate_table6(t6):
    consumo_en = t6.iloc[1:2,0:2]
    marcas_aprov = t6.iloc[0:1,1:]
    marcas_aprov = marcas_aprov.transpose()
    marcas_aprov[1] = 'Sí'
    return pd.concat([marcas_aprov,consumo_en]).reset_index(drop=True)
```

```python
def uglyt6(t6):
    nice_rows = []
    for i in range(len(t6)):
        possible_nice_row = deepcopy(t6.iloc[i:i+1,:].dropna(axis=1))
        if possible_nice_row.shape[1] == 2:
            nice_rows.append(possible_nice_row)
        else:
            nice_row = deepcopy(t6.iloc[i:i+1,:].reset_index(drop=True).transpose())
            nice_row[1] = 'Sí'
            nice_rows.append(nice_row)
    return pd.concat(nice_rows)
```
