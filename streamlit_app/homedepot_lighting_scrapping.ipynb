```python
def get_specs_table(link):
    table = pd.read_html(link)[-1]
    table_1 = copy.deepcopy(table.iloc[:,:2])
    table_2 = copy.deepcopy(table.iloc[:,2:]).rename(columns={2:0, 3:1})
    table = pd.concat([table_1,table_2],).reset_index(drop=True).T
    table.columns = table.iloc[0]
    table = table.loc[:,~table.columns.duplicated()].copy()
    return table.drop(0)
```

```python
def concat_all_tables(list_of_links):
    tables = []
    for link in list_of_links:
        try:
            tables.append(get_specs_table(link).reset_index(drop=True))
        except:
            print('bad link: {}'.format(link))
    return pd.concat(tables).reset_index(drop=True)
```
